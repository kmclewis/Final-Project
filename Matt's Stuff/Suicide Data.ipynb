```python
import pandas as pd
import numpy
from pandas import DataFrame
import csv
import glob
import os
from os import walk
```

```python
data_2015 = pd.read_csv('happy/2015.csv')
data_2016 = pd.read_csv('happy/2016.csv')
data_2017 = pd.read_csv('happy/2017.csv')
```

```python
data_2015.head(0)
```

```python
clean_2015 = data_2015.drop(['Region','Standard Error'], axis=1)
clean_2015.head(0)
```

```python
data_2016.head(0)
```

```python
clean_2016 = data_2016.drop(['Region', 'Lower Confidence Interval', 'Upper Confidence Interval'], axis =1)
```

```python
clean_2016.head(0)
```

```python
data_2017.head(0)
```

```python
clean_2017 = data_2017.drop(['Whisker.high', 'Whisker.low'], axis = 1)
clean_2017.head(0)
```

```python
clean_2017.rename(columns = {'Happiness.Rank' : 'Happiness Rank', 'Happiness.Score' :'Happiness Score',
                             'Economy..GDP.per.Capita.' : 'Economy (GDP per Capita)', 
                             'Health..Life.Expectancy.' : 'Health (Life Expectancy)', 
                             'Trust..Government.Corruption.' : 'Trust (Government Corruption)',
                             'Dystopia.Residual' : 'Dystopia Residual'}, inplace=True
                            )
```

```python
clean_2017.head(0)
```

```python
clean_2015.columns = ['2015 '+ str(col) for col in clean_2015.columns]
```

```python
clean_2015.rename(columns={'2015 Country': 'Country'}, inplace = True)
```

```python
clean_2015.head(0)
```

```python
clean_2016.columns = ['2016 '+ str(col) for col in clean_2016.columns] 
```

```python
clean_2016.rename(columns={'2016 Country': 'Country'}, inplace = True)
```

```python
clean_2016.head(0)
```

```python
clean_2017.columns = ['2017 '+ str(col) for col in clean_2017.columns]
```

```python
clean_2017.rename(columns={'2017 Country': 'Country'}, inplace = True)
```

```python
clean_2017.head(0)
```

```python
merged_df = pd.merge(clean_2015, clean_2016, on='Country', how='outer')
```

```python
merged_df.head()
```

```python
complete_df = pd.merge(merged_df, clean_2017,on='Country', how='outer')
```

```python
complete_df.head()
```

```python
dropped_df = complete_df.dropna()
```

```python
dropped_df
```

```python
dropped_df.to_csv('../happy/CountryHappinessRanks.csv', index=False)
```
